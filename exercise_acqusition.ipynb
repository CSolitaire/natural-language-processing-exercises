```python
# gets the page's title (the same text in the browser tab for a page, this is the <title> element
soup.title.string
```

```python
# is useful to print in case you want to see the HTML
soup.prettify()
```

```python
# find all the anchor tags, or whatever argument is specified.
soup.find_all("a")
```

```python
# finds the first matching element (.text provides the text of heading 1)
soup.find("h1")
```

```python
soup.find('h1').text
```

```python
# gets the text from within a matching piece of soup/HTML
soup.get_text()
```

```python
# method takes in a CSS selector as a string and returns all matching elements. super useful
soup.select(selector = 'div')
```
